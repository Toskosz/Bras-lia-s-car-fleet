```
percentage = data.copy(deep=True)
percentage = percentage.drop(data.columns[0], axis=1)
percentage
```

```
total_fleet_row = 0
for i in range(0, 11):
  for j in range(0, 9):
    percentage.iloc[i][j] = round((percentage.iloc[i][j]/data.iloc[total_fleet_row][0]) * 100, 2)
  total_fleet_row += 1
percentage
```
