## Limpeza e formatação de dados

Remoção das colunas "extras"

```
dados = dados.drop(dados.columns[[12, 13, 14]], axis=1)
dados
```

Remoção das linhas "extras"

```
dados = dados.drop(dados.index[[10, 11]])
dados
```

Renomeação da coluna 0 e retirada do index de numeração


```
dados.rename( columns={'Unnamed: 0':'Tipo de Veículo'}, inplace=True )
dados
```

```
dados = dados.set_index('Tipo de Veículo')
dados
```
